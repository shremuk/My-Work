```python
# How many crimes involve an arrest

%sql select description, count(arrest) as no_of_crimes_involved_as_arrest 
from CHICAGO_CRIME 
group by description, arrest having arrest = 'TRUE' 
```

```python
# Which unique types of crimes have been recorded at a GAS STATION?

%sql select DISTINCT primary_type as types_of_crime, location_description as location 
from chicago_crime 
where location_description = 'GAS STATION'
```

```python
# In the CENUS_DATA table list all Community Areas whose names start with the letter ‘B’.

%sql select community_area_name from chicago_socioeconomic where community_area_name like 'B%'
```

```python
# Which schools in Community Areas 10 to 15 are healthy school certified?

%sql select community_area_number, community_area_name, name_of_school, healthy_school_certified 
from chicago_school_reports 
where community_area_number between 10 and 15 and healthy_school_certified = 'Yes'
```

```python
# What is the average school Safety Score?

%sql select decimal(avg(safety_score),8,2) as average_safety_score from chicago_school_reports
```

```python
# List the top 5 Community Areas by average College Enrollment [number of students]

%sql select community_area_number, community_area_name, decimal(avg(college_enrollment),8,2) as avg_col_enrollment 
from chicago_school_reports group by community_area_number, community_area_name 
order by avg_col_enrollment desc 
limit 5
```

```python
# Use a sub-query to determine which Community Area has the least value for school Safety Score? Solution is given using 
# sub query by GROUP BY

%sql select community_area_number, community_area_name, safety_score 
from ( 
        select community_area_number, community_area_name, safety_score from chicago_school_reports 
        group by community_area_number, community_area_name, safety_score 
        order by min(safety_score) limit 1
     ) 
```

```python
# [Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score
# of 1

%sql select c.community_area_number, c.community_area_name, s.per_capita_income 
from chicago_school_reports as c, chicago_socioeconomic as s 
where c.community_area_number = s.community_area_number and safety_score = '1'
```
